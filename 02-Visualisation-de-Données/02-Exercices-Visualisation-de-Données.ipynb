# Exercices de Visualisation de Données

## Les données

Cet ensemble de données classique contient les prix et autres attributs de près de 54 000 diamants. C'est un ensemble de données idéal pour les débutants qui apprennent à travailler avec l'analyse et la visualisation de données.

#### Contenu

* *price* : prix en dollars US (\$326 -- \$18,823)

* *carat* : poids du diamant (0,2 -- 5,01)

* *cut* : qualité de la taille du diamant (Fair, Good, Very Good, Premium, Ideal) = (Moyen, Bon, Très bon, Supérieur, Idéal)

* *color* : couleur du diamant, de J (le pire) à D (le meilleur)

* *clarity* : mesure de la clarté du diamant (I1 (pire), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (meilleur))

* *x* : longueur en mm (0 -- 10,74)

* *y* : largeur en mm (0 -- 58,9)

* *z* : profondeur en mm (0 -- 31,8)
 
* *depth* : pourcentage total de la profondeur = z / moyenne (x, y) = 2 * z / (x + y) (43 -- 79)

* *table* : largeur du sommet du diamant par rapport au point le plus large (43 -- 95)

## Importations et Données

```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
diamonds = pd.read_csv('diamonds.csv')
```

```python
diamonds.head()
```

```python
diamonds.info()
```

## Tâches à effectuer

### Créer un diagramme de dispersion du prix (colonne price) par rapport au poids (colonne carat) comme indiqué ci-dessous

```python
sns.scatterplot(x='carat', y='price', data=diamonds)
```

### Utiliser le paramètre alpha et le paramètre edgecolor pour traiter le problème de chevauchement et le problème de marqueur de bord blanc.

```python
sns.scatterplot(x='carat', y='price', data=diamonds, alpha=0.1, color='blue', edgecolor='blue')
```

### Agrandir le graphique précédent

```python
plt.figure(figsize=(20,10))  # Taille de 20 pouces de large et 10 pouces de haut
sns.scatterplot(x='carat', y='price', data=diamonds, alpha=0.1, color='blue', edgecolor='blue')
```

### Créer un histogramme des prix (colonne price) comme indiqué ci-dessous

```python
sns.histplot(diamonds['price'], bins=30, color='blue')
```

### Créer un graphique de comptage des instances par qualité de taille du diamant comme indiqué ci-dessous

```python
sns.countplot(x='cut', data=diamonds, palette='tab10') 
```

### Créer un grand diagramme en boîtes montrant la répartition des prix par qualité de taille du diamant, comme indiqué ci-dessous

```python
sns.boxplot(x='cut', y='price', data=diamonds, palette='tab10')
```

### Défi ! Voyez si vous pouvez modifier l'ordre des boîtes à moustaches comme indiqué ci-dessous. N'hésitez pas à changer également la coloration, pour refléter cette relation entre les types de coupe.

```python
x=['Ideal', 'Premium', 'Good', 'Very Good', 'Fair']

sns.boxplot(x='cut', y='price', data=diamonds, palette='cool', order=x)
```
