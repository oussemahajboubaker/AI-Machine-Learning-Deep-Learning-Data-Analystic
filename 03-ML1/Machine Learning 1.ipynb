### Algorithme des k-voisins les plus proches

Cet algorithme tente de prédire la classe d’un échantillon de test en examinant les k échantillons d’apprentissage les plus proches (en distance) de l’échantillon de test. Par exemple, considérons le diagramme suivant dans lequel les points remplis représentent quatre exemples de classes: A, B, C et D. Nous utiliserons ces lettres comme noms de classe:


Nous voulons prédire les classes auxquelles appartiennent les nouveaux échantillons X, Y et Z. Supposons que nous souhaitons faire ces prédictions en utilisant les trois voisins les plus proches de chaque échantillon - trois est k dans l'algorithme des k-plus proches voisins:

• Les trois voisins les plus proches de l’échantillon X sont tous des points de classe D, nous prévoyons donc que la classe de X est D.

• Les trois voisins les plus proches de l’échantillon Y sont tous des points de classe B. Nous prévoyons donc que la classe de Y est B.

• Pour Z, le choix n'est pas aussi clair, car il apparaît entre les points B et C. Des trois voisins les plus proches, un est de classe B et deux sont de classe C. Dans l'algorithme des k-plus proches voisins, la classe avec le plus de «votes» l'emporte. Ainsi, sur la base de deux votes C pour un vote B, nous prédisons que la classe de Z est C. 

Choisir une valeur k impaire dans l'algorithme kNN évite les égalités en garantissant qu'il n'y a jamais un nombre égal de votes.

### Hyperparamètres et réglage des hyperparamètres
Dans l'apprentissage automatique, un modèle implémente un algorithme d'apprentissage automatique. Dans scikit-learn, les modèles sont appelés estimateurs. Il existe deux types de paramètres dans l'apprentissage automatique:

• ceux que l'estimateur calcule à mesure qu'il apprend des données que vous fournissez et

• ceux que vous spécifiez à l'avance lorsque vous créez l'objet d'estimation scikit-learn qui représente le modèle.

Les paramètres spécifiés à l'avance sont appelés hyperparamètres.
Dans l'algorithme des k voisins les plus proches, k est un hyperparamètre. Pour plus de simplicité, nous utiliserons les valeurs d'hyperparamètres par défaut de scikit-learn. Dans les études d'apprentissage automatique du monde réel, vous souhaiterez expérimenter différentes valeurs de k pour produire les meilleurs modèles possibles pour vos études. Ce processus est appelé réglage des hyperparamètres. Plus tard, nous utiliserons le réglage des hyperparamètres pour choisir la valeur de k qui permet à l'algorithme des k voisins les plus proches de faire les meilleures prédictions pour l'ensemble de données Digits. Scikit-learn dispose également de capacités de réglage automatisé des hyperparamètres.

Dans le rapport:

• la précision (precision ) est le nombre total de prédictions correctes pour un chiffre donné divisé par le nombre total de prédictions pour ce chiffre. Vous pouvez confirmer la précision en examinant chaque colonne de la matrice de confusion. Par exemple, si vous regardez l'index de colonne 7, vous verrez des 1 dans les lignes 3 et 4, indiquant que un 3 et un 4 ont été incorrectement classés comme 7 et un 45 dans la ligne 7 indiquant que les 45 images ont été correctement classées comme 7. La précision du chiffre 7 est donc 45/47 ou 0,96.

• rappel (recall) est le nombre total de prédictions correctes pour un chiffre donné divisé par le nombre total d'échantillons qui auraient dû être prédits comme ce chiffre. Vous pouvez confirmer le rappel en examinant chaque ligne de la matrice de confusion. Par exemple, si vous regardez l'index de ligne 8, vous verrez trois 1 et un 2 indiquant que certains 8 ont été incorrectement classés comme d'autres chiffres et un 39 indiquant que 39 images ont été correctement classées. Ainsi, le rappel pour le chiffre 8 est 39/44 ou 0,89.

• f1-score: il s'agit de la moyenne de la précision et du rappel.

• support: le nombre d'échantillons avec une valeur attendue donnée. Par exemple, 50 échantillons ont été étiquetés 4 et 38 échantillons ont été étiquetés 5.

Pour plus de détails sur les moyennes affichées au bas du rapport, voir:

### Validation croisée (K-Fold Cross-Validation)
La validation croisée K-fold vous permet d'utiliser toutes vos données à la fois pour l'entraînement et les tests, pour avoir une meilleure idée de la façon dont votre modèle fera des prédictions pour les nouvelles données en entraînant et en testant à plusieurs reprises le modèle avec différentes parties de l'ensemble de données. La validation croisée de K-fold divise l'ensemble de données en k plis de taille égale (ce k n'est pas lié à k dans l'algorithme des k-plus proches voisins). Vous entraînez ensuite à plusieurs reprises votre modèle avec des plis k - 1 et testez le modèle avec le pli restant. Par exemple, envisagez d'utiliser k = 10 avec des plis numérotés de 1 à 10. Avec 10 plis, nous ferions 10 cycles d'entraînement et de test successifs:

• Premièrement, nous nous entraînons avec les plis 1 à 9, puis nous effectuons un test avec le pli 10.

• Ensuite, nous nous entraînons avec les plis 1 à 8 et 10, puis nous testons avec le pli 9.

• Ensuite, nous nous entraînons avec les plis 1–7 et 9–10, puis testons avec le pli 8.

Ce cycle d'apprentissage et de test se poursuit jusqu'à ce que chaque pli ait été utilisé pour tester le modèle.

Scikit-learn fournit la classe KFold et la fonction cross_val_score (toutes deux dans le module sklearn.model_selection) pour vous aider à effectuer les cycles d'apprentissage et de test décrits ci-dessus. 

```python
# Classe KFold 
from sklearn.model_selection import KFold

kfold = KFold(n_splits=10, random_state=11, shuffle=True)
```

```python
# Utilisation de l'objet KFold avec la fonction cross_val_score
from sklearn.model_selection import cross_val_score

scores = cross_val_score(estimator=knn, X=digits.data, 
     y=digits.target, cv=kfold)
scores
```

```python
print(f'Mean accuracy: {scores.mean():.2%}')

print(f'Accuracy standard deviation: {scores.std():.2%}')
```

Bien que nous n’ayons pas étudié les estimateurs SVC et GaussianNB, scikit-learn vous permet néanmoins de les tester facilement en utilisant leurs paramètres par défaut.

Pour déterminer la meilleure valeur de k dans l’algorithme kNN, essayez différentes valeurs de k, puis comparez les performances de l’estimateur avec chacune. Nous pouvons le faire en utilisant des techniques similaires à la comparaison d'estimateurs. La boucle suivante crée des KNeighborsClassifiers avec des valeurs k impaires de 1 à 19 et effectue une validation croisée de k fois sur chacun. Vous pouvez également voir que la précision a tendance à diminuer pour des valeurs k plus élevées:

```python
for k in range(1, 20, 2):
     kfold = KFold(n_splits=10, random_state=11, shuffle=True)
     knn = KNeighborsClassifier(n_neighbors=k)
     scores = cross_val_score(estimator=knn, 
         X=digits.data, y=digits.target, cv=kfold)
     print(f'k={k:<2}; mean accuracy={scores.mean():.2%}; ' +
           f'standard deviation={scores.std():.2%}')
```

## Overfitting/Underfitting

Lors de la création d'un modèle, un objectif clé est de s'assurer qu'il est capable de faire des prédictions précises pour les données qu'il n'a pas encore vues. Deux problèmes courants qui empêchent des prédictions précises sont le surajustement et le sous-ajustement:

• Un sous-ajustement se produit lorsqu'un modèle est trop simple pour faire des prédictions, sur la base de ses données d'apprentissage. Par exemple, vous pouvez utiliser un modèle linéaire, tel qu'une régression linéaire simple, alors qu'en fait, le problème nécessite vraiment un modèle non linéaire. Par exemple, les températures varient considérablement au cours des quatre saisons. Si vous essayez de créer un modèle général capable de prédire les températures tout au long de l’année, un simple modèle de régression linéaire sous-adaptera les données.

• Le surajustement se produit lorsque votre modèle est trop complexe. Le cas le plus extrême serait un modèle qui mémorise ses données d'entraînement. Cela peut être acceptable si vos nouvelles données ressemblent exactement à vos données d'entraînement, mais ce n'est généralement pas le cas. Lorsque vous faites des prédictions avec un modèle de surajustement, de nouvelles données qui correspondent aux données d'entraînement produiront des prédictions parfaites, mais le modèle ne saura pas que faire des données qu'il n'a jamais vues.

Pour plus d'informations sur le sous-ajustement et le sur-ajustement, voir

• https://en.wikipedia.org/wiki/Overfitting

• https://machinelearningmastery.com/overfitting-and-underfittingwith-machine-learning-algorithms/
