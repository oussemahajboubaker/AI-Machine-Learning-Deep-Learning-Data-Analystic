### Algorithme des k-voisins les plus proches

Cet algorithme tente de prdire la classe dun chantillon de test en examinant les k chantillons dapprentissage les plus proches (en distance) de lchantillon de test. Par exemple, considrons le diagramme suivant dans lequel les points remplis reprsentent quatre exemples de classes: A, B, C et D. Nous utiliserons ces lettres comme noms de classe:


Nous voulons prdire les classes auxquelles appartiennent les nouveaux chantillons X, Y et Z. Supposons que nous souhaitons faire ces prdictions en utilisant les trois voisins les plus proches de chaque chantillon - trois est k dans l'algorithme des k-plus proches voisins:

 Les trois voisins les plus proches de lchantillon X sont tous des points de classe D, nous prvoyons donc que la classe de X est D.

 Les trois voisins les plus proches de lchantillon Y sont tous des points de classe B. Nous prvoyons donc que la classe de Y est B.

 Pour Z, le choix n'est pas aussi clair, car il apparat entre les points B et C. Des trois voisins les plus proches, un est de classe B et deux sont de classe C. Dans l'algorithme des k-plus proches voisins, la classe avec le plus de votes l'emporte. Ainsi, sur la base de deux votes C pour un vote B, nous prdisons que la classe de Z est C. 

Choisir une valeur k impaire dans l'algorithme kNN vite les galits en garantissant qu'il n'y a jamais un nombre gal de votes.

### Validation croise (K-Fold Cross-Validation)
La validation croise K-fold vous permet d'utiliser toutes vos donnes  la fois pour l'entranement et les tests, pour avoir une meilleure ide de la faon dont votre modle fera des prdictions pour les nouvelles donnes en entranant et en testant  plusieurs reprises le modle avec diffrentes parties de l'ensemble de donnes. La validation croise de K-fold divise l'ensemble de donnes en k plis de taille gale (ce k n'est pas li  k dans l'algorithme des k-plus proches voisins). Vous entranez ensuite  plusieurs reprises votre modle avec des plis k - 1 et testez le modle avec le pli restant. Par exemple, envisagez d'utiliser k = 10 avec des plis numrots de 1  10. Avec 10 plis, nous ferions 10 cycles d'entranement et de test successifs:

 Premirement, nous nous entranons avec les plis 1  9, puis nous effectuons un test avec le pli 10.

 Ensuite, nous nous entranons avec les plis 1  8 et 10, puis nous testons avec le pli 9.

 Ensuite, nous nous entranons avec les plis 17 et 910, puis testons avec le pli 8.

Ce cycle d'apprentissage et de test se poursuit jusqu' ce que chaque pli ait t utilis pour tester le modle.

Scikit-learn fournit la classe KFold et la fonction cross_val_score (toutes deux dans le module sklearn.model_selection) pour vous aider  effectuer les cycles d'apprentissage et de test dcrits ci-dessus. 

```python
# Classe KFold 
from sklearn.model_selection import KFold

kfold = KFold(n_splits=10, random_state=11, shuffle=True)
```

```python
# Utilisation de l'objet KFold avec la fonction cross_val_score
from sklearn.model_selection import cross_val_score

scores = cross_val_score(estimator=knn, X=digits.data, 
     y=digits.target, cv=kfold)
scores
```

```python
print(f'Mean accuracy: {scores.mean():.2%}')

print(f'Accuracy standard deviation: {scores.std():.2%}')
```

## Overfitting/Underfitting

Lors de la cration d'un modle, un objectif cl est de s'assurer qu'il est capable de faire des prdictions prcises pour les donnes qu'il n'a pas encore vues. Deux problmes courants qui empchent des prdictions prcises sont le surajustement et le sous-ajustement:

 Un sous-ajustement se produit lorsqu'un modle est trop simple pour faire des prdictions, sur la base de ses donnes d'apprentissage. Par exemple, vous pouvez utiliser un modle linaire, tel qu'une rgression linaire simple, alors qu'en fait, le problme ncessite vraiment un modle non linaire. Par exemple, les tempratures varient considrablement au cours des quatre saisons. Si vous essayez de crer un modle gnral capable de prdire les tempratures tout au long de lanne, un simple modle de rgression linaire sous-adaptera les donnes.

 Le surajustement se produit lorsque votre modle est trop complexe. Le cas le plus extrme serait un modle qui mmorise ses donnes d'entranement. Cela peut tre acceptable si vos nouvelles donnes ressemblent exactement  vos donnes d'entranement, mais ce n'est gnralement pas le cas. Lorsque vous faites des prdictions avec un modle de surajustement, de nouvelles donnes qui correspondent aux donnes d'entranement produiront des prdictions parfaites, mais le modle ne saura pas que faire des donnes qu'il n'a jamais vues.

Pour plus d'informations sur le sous-ajustement et le sur-ajustement, voir

 https://en.wikipedia.org/wiki/Overfitting

 https://machinelearningmastery.com/overfitting-and-underfittingwith-machine-learning-algorithms/
